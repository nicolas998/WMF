# Load packages

```python
%matplotlib inline 
import pylab as pl 
from wmf import wmf 
import numpy as np 
import pandas as pd 
import os
import osgeo
from wmf import ghost_topo as go 
```

# Define the watershed in the region

```python
DEM, DIR,epsg  = wmf.dem_process('../data/dem.tif', dxp = 96, noData = -9999)
```

```python
pl.figure(figsize=(9,8))
pl.imshow(DIR.T)
```

## Guide streamflow 

This element helps to correct the watershed outlet coordinates 

```python
st = wmf.Stream(352378,4605633, DEM, DIR)
st.Save_Stream2Map('../data/guide_stream.shp', EPSG = int(epsg))
```

## Watershed element in WMF

```python
cu = wmf.SimuBasin(340517,4589890, DEM, DIR, stream=st, threshold=50)
cu.Save_Basin2Map('../data/watershed.shp', dx = 96, EPSG = int(epsg), )
```

```python
cu.plot_basin(shape_path='../data/dem_network2.shp', 
    shape_width=3)
```

# Ghost pre-processing

```python
#cu = wmf.SimuBasin(path='../data/basin_example.nc')
gp = go.ghost_preprocess(cu, path_dem='../data/dem.tif', 
    seg_threshold=700,
    seg_point_distance=75)
```

```python
gp.get_segments_topology()
gp.get_mesh_river_points(dist = 100,clean_close_points=True,
                         min_river2river_distance=50)
gp.get_mesh_grid_points(mesh_spaces=5,
                        border_iter=2,
                        clean_with_river=True,
                        min_dem2river_distance=50)

gp.get_voronoi_polygons()
gp.define_polygons_topology()

#Mesh file and shp (map)
gp.write_mesh_file(path = '../data/example.mesh', 
                   shp_path = '../data/example_mesh.shp')
#River file and shp (map)
gp.write_river_file(path = '../data/exampl.riv', 
                    shp_path = '../data/example_riv.shp')
```
